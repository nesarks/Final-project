# Data Collection

## Simulated user data

```python
import random
import pandas as pd
```

```python
# Defining lists of possible vallues
financial_goals = [random.randint(10000, 100000) for _ in range(1000)]  # Random goals between $10K and $100K
risk_tolerance_levels = ['Low', 'Medium', 'High']
investment_horizons = ['Short', 'Medium', 'Long']
```

```python
# Generating synthetic user profiles
users = []
for _ in range(1000):  # Generating 1000 user profiles
    user_profile = {
        'Financial Goal': random.choice(financial_goals),
        'Risk Tolerance': random.choice(risk_tolerance_levels),
        'Investment Horizon': random.choice(investment_horizons)
    }
    users.append(user_profile)
```

```python
# converting the list of users into a df
user_df = pd.DataFrame(users)
user_df.head()
```

```python
user_df.to_csv('user_data.csv', index=False)
```

# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
# visualizing user data
plt.figure(figsize=(10, 5))
plt.hist(user_df['Financial Goal'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of User Financial Goals')
plt.xlabel('Financial Goal ($)')
plt.ylabel('Number of Users')
plt.show()
```

```python
# visualizing risk tolerance levels
plt.figure(figsize=(8, 5))
user_df['Risk Tolerance'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title('Distribution of User Risk Tolerance Levels')
plt.xlabel('Risk Tolerance Level')
plt.ylabel('Number of Users')
plt.show()
```

```python
# visualizing investment horizon distribution
plt.figure(figsize=(8, 5))
user_df['Investment Horizon'].value_counts().plot(kind='bar', color='purple', edgecolor='black')
plt.title('Distribution of User Investment Horizons')
plt.xlabel('Investment Horizon')
plt.ylabel('Number of Users')
plt.show()
```
